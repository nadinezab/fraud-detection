# Credit Card Fraud Detection

### Introduction


This project focuses on leveraging diverse predictive models to evaluate their accuracy in discerning between legitimate transactions and instances of credit card fraud. The dataset employed for this analysis adheres to privacy standards, with features appropriately scaled and names anonymized to safeguard sensitive information. Despite these privacy measures, the project aims to delve into crucial aspects of the dataset, shedding light on patterns and characteristics that can contribute to the development of robust fraud detection mechanisms without compromising individual privacy. Through this exploration, the goal is to enhance our understanding of the underlying dynamics of credit card transactions and fortify the efficacy of fraud detection methodologies.

### Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Data/creditcard.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().any().sum()
```

We see there are no missing values in the dataset.

```python
print('Legitimate Transactions', round(df['Class'].value_counts()[0]/len(df) * 100,2), '% of the dataset')
print('Fraudulent Transactions', round(df['Class'].value_counts()[1]/len(df) * 100,2), '% of the dataset')
```

We notice the dataset is heavily imbalanced, with the majority of transactions being legitimate. This is important to take into consideration as otherwise our algorithm might achieve high accuracy simply by predicting the majority class.

```python
sns.countplot(x='Class', data=df)
plt.title('Class Distribution', fontsize=14)
plt.xticks(ticks = [0,1], labels=['Legitimate', 'Fraud'])
plt.plot()
```

Looking at the descriptive statistics above, we see thatt most of the data has been scaled. Let us scale columns amount and time too.

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

df['scaled_amount'] = std_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['scaled_time'] = std_scaler.fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
from scipy.stats import norm

f, (ax1, ax2, ax3) = plt.subplots(1,10, figsize=(20, 6))

v1_fraud_dist = df['V1'].loc[df['Class'] == 1].values
sns.histplot(v1_fraud_dist,ax=ax1, color='#FB8861')
ax1.set_title('V11 Distribution \n (Fraud Transactions)', fontsize=14)

v12_fraud_dist = df['V12'].loc[df['Class'] == 1].values
sns.histplot(v12_fraud_dist,ax=ax2, color='#56F9BB')
ax2.set_title('V12 Distribution \n (Fraud Transactions)', fontsize=14)


v10_fraud_dist = df['V10'].loc[df['Class'] == 1].values
sns.histplot(v10_fraud_dist,ax=ax3, color='#C5B3F9')
ax3.set_title('V10 Distribution \n (Fraud Transactions)', fontsize=14)

plt.show()
```

```python
palette = sns.color_palette("husl", 30)
palette[0]
```

```python

f, axes = plt.subplots(nrows=10,ncols=3, figsize=(20, 60))

axes = axes.flatten()

for i in range(0,28):
    sns.histplot(df[f'V{i+1}'].loc[df['Class'] == 1].values,ax=axes[i], color=palette[i])
    axes[i].set_title(f'V{i+1} Distribution \n (Fraud Transactions)', fontsize=14)
    
sns.histplot(df['scaled_amount'].loc[df['Class'] == 1].values,ax=axes[28])
axes[28].set_title('Scaled Amount Distribution \n (Fraud Transactions)', fontsize=14)

sns.histplot(df['scaled_time'].loc[df['Class'] == 1].values,ax=axes[29])
axes[29].set_title('Scaled Time Distribution \n (Fraud Transactions)', fontsize=14)

plt.show()
```
